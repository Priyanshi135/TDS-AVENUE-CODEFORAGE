13. Classify Movies based on Ratings

```python
data.Rating
```

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['rating_category']=data['Rating'].apply(rating)
```
